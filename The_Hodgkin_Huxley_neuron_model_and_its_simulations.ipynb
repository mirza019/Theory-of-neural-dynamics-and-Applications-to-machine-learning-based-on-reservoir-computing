```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
Cm = 1.0
gK = 36.0
gNa = 120.0
gL = 0.3
VK = -12.0
VNa = 115.0
VL = 10.0
```

```python

t_span = [0, 200]
t_eval = np.linspace(0, 200, 500)
```

```python
def hodgkin_huxley(t, y, Iext):
    V, n, m, h = y
    dVdt = (1/Cm) * (gK * n**4 * (VK - V) + gNa * m**3 * h * (VNa - V) + gL * (VL - V) + Iext)
    an = 0.01 * (10 - V) / (np.exp((10 - V) / 10) - 1) if V != 10 else 0.1
    bn = ( np.exp(-V / 80))/8
    am = 0.1 * (25 - V) / (np.exp((25 - V) / 10) - 1) if V != 25 else 1
    bm = 4 * np.exp(-V / 18)
    ah = 0.07 * np.exp(-V / 20)
    bh = 1 / (1 + np.exp((30 - V) / 10))
    dndt = an * (1 - n) - bn * n
    dmdt = am * (1 - m) - bm * m
    dhdt = ah * (1 - h) - bh * h
    return [dVdt, dndt, dmdt, dhdt]
```

```python
#Problem 1
Iext = 5.2
initial_conditions = [0, 0.35, 0.06, 0.6]


sol = solve_ivp(hodgkin_huxley, t_span, initial_conditions, args=(Iext,),
                t_eval=t_eval, method='RK45')
# 1st Plotting
fig, axs = plt.subplots(4, 1, figsize=(8, 12), sharex=True)
axs[0].plot(sol.t, sol.y[0], label='V(t)')
axs[1].plot(sol.t, sol.y[1], label='n(t)')
axs[2].plot(sol.t, sol.y[2], label='m(t)')
axs[3].plot(sol.t, sol.y[3], label='h(t)')
for ax in axs:
    ax.legend()
    ax.set_xlabel('Time (ms)')
axs[0].set_ylabel('Membrane potential (mV)')
axs[1].set_ylabel('Potassium gating variable')
axs[2].set_ylabel('Sodium activation variable')
axs[3].set_ylabel('Sodium inactivation variable')
plt.suptitle("Hodgkin-Huxley Neuron Dynamics for Problem 1")
plt.show()

# 2nd Plotting
plt.figure(figsize=(10, 8))
plt.plot(sol.t, sol.y[0], label='V(t)')
plt.plot(sol.t, sol.y[1], label='n(t)')
plt.plot(sol.t, sol.y[2], label='m(t)')
plt.plot(sol.t, sol.y[3], label='h(t)')
plt.title('Hodgkin-Huxley Neuron Model Problem 1')
plt.xlabel('Time (ms)')
plt.ylabel('Variables')
plt.legend()
plt.show()
```
